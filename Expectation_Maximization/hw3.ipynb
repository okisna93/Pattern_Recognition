```python
#ISMET OKAN CELIK CWID:10472265
import numpy as np
import scipy.io as io
from scipy.stats import multivariate_normal

hw3_1 = io.loadmat("hw3.mat")['hw3_1']
hw3_1=np.array(hw3_1)
```

```python
#First Iteration
r=0.5
group1_initial=np.array(hw3_1[:,1:50])
group2_initial=np.array(hw3_1[:,50:100])

mu1_initial=np.average(group1_initial,axis=1)
mu2_initial=np.average(group2_initial,axis=1)

print('mu1 = ',mu1_initial,'\n')
print('mu2 =',mu2_initial,'\n')

coVar1_initial=np.cov(group1_initial)
coVar2_initial=np.cov(group2_initial)

print('CoVarience-1 Initial = \n',coVar1_initial,'\n')
print('CoVarience-2 Initial = \n',coVar2_initial,'\n')
```

```python
def Grouping_Samples(mu1,mu2,coVar1,coVar2):
    p1x=[]
    p2x=[]

    group1x=[]
    group1y=[]

    group2x=[]
    group2y=[]

    for i in range(len(hw3_1[0])):
        p1x_1=multivariate_normal.pdf((hw3_1[0][i],hw3_1[1][i]),mu1,coVar1)
        p2x_1=multivariate_normal.pdf((hw3_1[0][i],hw3_1[1][i]),mu2,coVar2)
        p1x.append(p1x_1)
        p2x.append(p2x_1)

        if p1x_1>p2x_1:
            group1x.append(hw3_1[0][i])
            group1y.append(hw3_1[1][i])
        else:
            group2x.append(hw3_1[0][i])
            group2y.append(hw3_1[1][i])

    group1=[group1x,group1y]
    group1=np.array(group1)

    group2=[group2x,group2y]
    group2=np.array(group2)

    print('1.Group - After Iteration \n','1.Group Size =', len(group1x),'\n',group1,'\n')
    print('2.Group - After Iteration \n','2.Group Size =',len(group2x),'\n',group2) 
    return group1, group2

group1_1,group2_1=Grouping_Samples(mu1_initial,mu2_initial,coVar1_initial,coVar2_initial)
```

```python
#Second Iteration

def mu_coVar(group1,group2):
    mu1=np.average(group1,axis=1)
    mu2=np.average(group2,axis=1)

    print('mu1 and mu2 After Iteration')
    print('mu1 = ',mu1)
    print('mu2 =',mu2,'\n')

    coVar1=np.cov(group1)
    coVar2=np.cov(group2)

    print('CoVarience-1 After Iteration = \n',coVar1,'\n')
    print('CoVarience-2 After Iteration = \n',coVar2,'\n')
    return mu1,mu2,coVar1,coVar2

mu1_2,mu2_2,coVar1_2,coVar2_2=mu_coVar(group1_1,group2_1)
group1_2,group2_2=Grouping_Samples(mu1_2,mu2_2,coVar1_2,coVar2_2)
```

```python
#Third Iteration
mu1_3,mu2_3,coVar1_3,coVar2_3=mu_coVar(group1_1,group2_1)
group1_3,group2_3=Grouping_Samples(mu1_3,mu2_3,coVar1_3,coVar2_3)
```

```python
#Fourth Iteration
mu1_4,mu2_4,coVar1_4,coVar2_4=mu_coVar(group1_3,group2_3)
group1_4,group2_4=Grouping_Samples(mu1_4,mu2_4,coVar1_4,coVar2_4)
```

```python
#Fifth Iteration
mu1_5,mu2_5,coVar1_5,coVar2_5=mu_coVar(group1_4,group2_4)
group1_5,group2_5=Grouping_Samples(mu1_5,mu2_5,coVar1_5,coVar2_5)

#Estimated Ro
r=len(group1_5[0])/len(hw3_1[0])
print('\n Estimated Ro Number=',r)
```
