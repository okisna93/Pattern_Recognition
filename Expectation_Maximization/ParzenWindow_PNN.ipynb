```python
#ISMET OKAN CELIK CWID:10472265
# Question-2 
import numpy as np
import scipy.io as io
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

hw3_2_1 = io.loadmat("hw3.mat")['hw3_2_1']
hw3_2_1=np.array(hw3_2_1)
hw3_2_2 = io.loadmat("hw3.mat")['hw3_2_2']
hw3_2_2=np.array(hw3_2_2)
```

```python
h1=2
n=100
x=np.array([np.arange(-4,8,0.1),np.arange(-4,8,0.1)])
k=1/np.sqrt(n)
hn=h1/np.sqrt(n)
```

```python

window_func_1=[]     
window_func_2=[]

for k in range(len(x[0])):
    px_list_1=[]
    px_list_2=[]
    for i in range(len(hw3_2_1[0])):
        a=np.array([[x[0][k]],[x[1][k]]])
        data1=np.array([[hw3_2_1[0][i]],[hw3_2_1[1][i]]])
        data2=np.array([[hw3_2_2[0][i]],[hw3_2_2[1][i]]])
        
        # Eucledian distance calculation
        u1=np.linalg.norm(a-data1)/hn
        u2=np.linalg.norm(a-data2)/hn
        
        window_func1=(1/(np.sqrt(2*np.pi)))*np.exp(-1*np.power(u1,2)/2)
        window_func2=(1/(np.sqrt(2*np.pi)))*np.exp(-1*np.power(u2,2)/2)
        window_func_1.append((1/hn)*window_func1)
        window_func_2.append((1/hn)*window_func2)

N = 100
subList_window1 = [window_func_1[i:i+N] for i in range(0, len(window_func_1), N)]
subList_window2 = [window_func_2[i:i+N] for i in range(0, len(window_func_2), N)]

'''
After spliting all window function calculation to small subsets for ever iteration,
we are going to calculate the p(x) function for two different data sets
'''

p1x_1=[]
p2x_1=[]
for i in range(len(subList_window1)):
    a=sum(subList_window1[i])
    b=sum(subList_window2[i])
    p1x_1.append((1/n)*a)
    p2x_1.append((1/n)*b)
       
p1x_1=np.array(p1x_1)
p2x_1=np.array(p2x_1)

print('Number of Elements in the list p1x =', len(p1x_1))
print('\n Summation of the p1x Value for every x Location vector \n',p1x_1)

print('\n Number of Elements in the list p2x =', len(p2x_1))
print('\n Summation of the p2x Value for every x Location vector \n',p1x_1)
```

```python
# Probability Value Visulazation for class-1
X = np.array(np.arange(-4,8,0.1))
Y = np.array(np.arange(-4,8,0.1))
Z =p1x_1

x = np.reshape(X, (10, 12))
y = np.reshape(Y, (10, 12))
z = np.reshape(Z, (10, 12))


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z,rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
#Probability Value Visulazation for class-2
X = np.array(np.arange(-4,8,0.1))
Y = np.array(np.arange(-4,8,0.1))
Z =p2x_1

x = np.reshape(X, (10, 12))
y = np.reshape(Y, (10, 12))
z = np.reshape(Z, (10, 12))


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z,rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
#Classification of given x=[1,-2]'
x_input=np.array([[1],[-2]])
px_input_1=[]
px_input_2=[]
for i in range(len(hw3_2_1[0])):
    a=np.array([[x_input[0]],[x_input[1]]])
    data1=np.array([[hw3_2_1[0][i]],[hw3_2_1[1][i]]])
    data2=np.array([[hw3_2_2[0][i]],[hw3_2_2[1][i]]])
        
    # 1-D with eucledian distance calculation
        
    u1=np.linalg.norm(a-data1)/hn
    u2=np.linalg.norm(a-data2)/hn
    window_func1=(1/(np.sqrt(2*np.pi)))*np.exp(-1*np.power(u1,2)/2)
    window_func2=(1/(np.sqrt(2*np.pi)))*np.exp(-1*np.power(u2,2)/2)
    px_input_1.append((1/hn)*window_func1)
    px_input_2.append((1/hn)*window_func2)
    
print('p1x for class-1',sum(px_input_1))
print('p2x for class-2',sum(px_input_2))
print('x=[1,-2] belongs to class-1')
```

```python
# Probabilistic Neural Network (PNN)
x_input=np.array([[1],[-2]])
w1=hw3_2_1
w2=hw3_2_2
sigma=0.2

fk1=[]
fk2=[]

for i in range(len(w1[0])):
    # Calculation of Net Activation
    netk1=np.dot(np.transpose([[w1[0][i]],[w1[1][i]]]),x_input)
    netk2=np.dot(np.transpose([[w2[0][i]],[w2[1][i]]]),x_input)

    # Calculation of Active Function
    fnetk1=np.exp((netk1-1)/(np.power(sigma,2)))
    fnetk2=np.exp((netk2-1)/(np.power(sigma,2)))
    fk1.append(fnetk1)
    fk2.append(fnetk2)
    
print('Value of fknet1 =',sum(fk1))
print('Value of fknet2 =',sum(fk2))
print('Comparing the two value fknet1 > fknet2 , we can say that input value x belongs to class-1')
```
